```python
import decimal
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def napr(x,y):
    result = (y + 2.5*(x-1)-4.5)/x
    return result
```

```python
def float_range(start, stop, step):
  while start < stop:
    yield float(start)
    start += decimal.Decimal(step)
```

```python
dots = [[1],[1]]
```

```python
def f_y(x,k,x0,y0):
    result = (k*(x-x0)+y0)
    return result
```

```python
def euler(n_x,x,y):
    global x0,y0
    k = napr(x,y)
    y2 = k*(n_x-x)+y
    return y2
```

```python
def get_dots():
    for i in range(len(dots[0])):
        yield (dots[0][i],dots[1][i])
```

```python
def build_graf(STEP):
    global dots
    for dot in float_range(1,3+STEP,STEP):
        y = euler(dot,dots[0][-1],dots[1][-1])
        dots[0].append(dot)
        dots[1].append(y)
    del dots[0][0]
    del dots[1][0]
    plt.plot(dots[0],dots[1])
    plt.show()
    print(dots[1][dots[0].index(3.0)])
    dots = [[1],[3]]
```
